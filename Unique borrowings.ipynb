```python
import spacy
import pandas as pd
from re import T
import nltk
import csv
from nltk import ngrams
```

```python
#DF with English, original Spanish translation and machine translated in Spanish
df = pd.read_excel("/content/drive/MyDrive/MT assignment/ARCHIVOS FINALES PARA ANALIZAR/loans_all_clean.xlsx", names=['sentence_en', 'sentence_es', 'sentence_es_mt', 'borrowings_raw_es', 'borrowings_clean_es','borrowings_clean_mt','borrowings_raw_mt'])
```

```python
from collections import defaultdict
def unique_borrowings(df, ver):

  unique_borrowings_dict = defaultdict(int)
  nlp = spacy.load("es_core_news_sm", disable=["parser", "ner"])
  for idx, row in df.iterrows():
    doc = nlp(row['borrowings_clean_'+ver])
    for token in doc:
      if token.text != 'O':
        unique_borrowings_dict[token.lemma_] +=1
  return unique_borrowings_dict


unique_borrowings_1 = unique_borrowings(df,'es')
unique_borrowings_2 = unique_borrowings(df,'mt')
```

```python
keys = []
for key,value in unique_borrowings_1.items():
  keys.append(key)
  print(f"{key}:{value}")

print(len(keys))
```
