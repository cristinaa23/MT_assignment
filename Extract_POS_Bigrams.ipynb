#POS TAGGING + BIGRAMS

```python
nlp = spacy.load("en_core_web_trf")
nlp_es = spacy.load("es_core_news_sm")
```

```python
def obtener_bigramas(oracion, nlp):
    doc = nlp(oracion)
    bigramas = []
    for i in range(len(doc) - 1):
        bigramas.append((doc[i].text, doc[i+1].text, doc[i].pos_, doc[i+1].pos_))
    return bigramas
```

```python
#SAMPLE 1

# Crear una lista con las oraciones y sus bigramas
lista_oraciones_bigramas = []
for i in range(len(sents_en1)):
    en_bigramas = obtener_bigramas(sents_en1[i], nlp)
    og_es_bigramas = obtener_bigramas(sents_og_es1[i], nlp_es)
    mt_es_bigramas = obtener_bigramas(sents_mt_es1[i], nlp_es)
    lista_oraciones_bigramas.append((sents_en1[i], sents_og_es1[i], sents_mt_es1[i], en_bigramas, og_es_bigramas, mt_es_bigramas))

# Save results in CSV file
with open('/content/drive/MyDrive/MT assignment/temporales_POS/sents_bigrams1.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['En_sents', 'Es_sents', 'Es_sents_MT', 'En_bigramas', 'Es_bigramas', 'Es_bigramas_MT'])
    for oracion in lista_oraciones_bigramas:
        writer.writerow(oracion)
```

# EXTRACTING VERB + ADP

```python
df = pd.read_csv('/content/drive/MyDrive/MT assignment/temporales_POS/sents_bigrams_all.csv')
import ast

def bigrams(df, column_name):
    for idx, row in df.iterrows():
        pos_tag_list = ast.literal_eval(row[column_name])
        bigram_list = []
        for i in range(len(pos_tag_list)-1):
            if (pos_tag_list[i][2] == 'VERB' or pos_tag_list[i][2] == 'AUX') and pos_tag_list[i+1][2] == 'ADP':
                bigram_list.append(pos_tag_list[i])
        if len(bigram_list) > 0:
            df.at[idx, column_name +'Clean'] = str(bigram_list)
        else:
            df.at[idx, column_name +'Clean'] = 'O'
    return df
   # if len(v_adp_tuples) > 0:
   #     df.at[idx, 'new_column'] = v_adp_tuples
   # else:
   #     df.at[idx, 'new_column'] = 'O'


column_names = ['En_bigramas', 'Es_bigramas', 'Es_bigramas_MT']

for column_name in column_names:
  df = bigrams(df, column_name)
  df = bigrams(df, column_name)


# Apply the function to the "text" column of the dataframe

df.to_csv('/content/drive/MyDrive/MT assignment/POS_analysis_final.csv', index=False)
#print(df['En_bigramas'][0][3])

display(df)
```
