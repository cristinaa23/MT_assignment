```python
import pandas as pd
import spacy
df = pd.read_csv("/content/drive/MyDrive/MT assignment/mt_translated_all.csv", names=['sentence_en', 'sentence_es', 'sentence_mt'])
```

```python
display(df)
```

```python
def adverb_finishing_in_ly(df, ver):
  spacy.require_gpu()
  count = 0
  if ver == 'en':
    nlp = spacy.load("en_core_web_trf", disable=["parser", "ner"])
    for idx, row in df.iterrows():
        adverb_list = []
        doc = nlp(row['sentence_'+ver])
        adverb = ""
        for token in doc:
            if token.pos_ == "ADV" and token.text.endswith("ly"):
                adverb = token.text
                adverb_list.append(adverb)
                count += 1
        if len(adverb_list) > 0:
            df.at[idx, ver +'Adv'] = adverb_list
        else:
            df.at[idx, ver +'Adv'] = 'O'
  elif ver == 'es' or ver== "mt": 
      nlp = spacy.load("es_dep_news_trf", disable=["parser", "ner"])
      for idx, row in df.iterrows():
        adverb_list = []
        doc = nlp(row['sentence_'+ver])
        adverb = ""
        for token in doc:
            if token.pos_ == "ADV" and token.text.endswith("mente"):
                adverb = token.text
                adverb_list.append(adverb)
                count += 1
        if len(adverb_list) > 0:
            df.at[idx, ver +'Adv'] = adverb_list
        else:
            df.at[idx, ver +'Adv'] = 'O'
  print(f"Number of adverbs in sentence_{ver}: {count}")
  return df

version = ['en', 'es', 'mt']
for ver in version:
  new_df = adverb_finishing_in_ly(df, ver)



```
